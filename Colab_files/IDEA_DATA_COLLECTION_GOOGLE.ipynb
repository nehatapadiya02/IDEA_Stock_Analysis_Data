```python
import re

def splitSourceAndTitle(title):
# Splitting the string using regex
  split_result = re.split(r' - ', title)

  # Extract the last part and all before it
  before_last = ' - '.join(split_result[:-1])  # All before the last '-'
  last_part = split_result[-1]  # The last part

  return before_last, last_part
```

```python
news_data['title_new'], news_data['source'] = zip(*news_data['title'].apply(splitSourceAndTitle))
```

```python
news_data.drop_duplicates(subset = [ "Stock_Name","title_new"],
                     keep = 'first', inplace = True)
```

```python
news_data["title_new"] = news_data["title_new"].str.lower().str.replace(r"[^\w\s]", "", regex=True)
news_data["description"] = news_data["description"].str.lower().str.replace(r"[^\w\s]", "", regex=True)
```

```python
indian_holidays = holidays.India(years=[2024]).keys()
indian_business_day = pd.offsets.CustomBusinessDay(holidays=list(indian_holidays))

# Create a set of holiday dates using NumPy's datetime64
indian_holidays_set = set(np.datetime64(date) for date in indian_holidays)
def next_indian_business_day(date):
    # Convert input date to numpy datetime64 if it isn't already
    date = np.datetime64(date)

    # Loop until a valid business day is found
    while date in indian_holidays_set or pd.Timestamp(date).weekday() in [5, 6]:  # 5 = Saturday, 6 = Sunday
        # Add one day
        date = date + np.timedelta64(1, 'D')

    return date
```

```python
news_data["Adjusted_Date"] = news_data["date_formatted"].apply(
    lambda x: next_indian_business_day(x)
)
```
